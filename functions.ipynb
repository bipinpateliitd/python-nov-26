 * **kwargs (Keyword Arguments)
âœ… Meaning

**kwargs allows a function to accept any number of keyword arguments.

ðŸ“Œ Python stores them in a dictionary.

```python
def student_details(**dict_inputs):
    print(dict_inputs)

```

```python
student_details(name="Suresh", age=25, course="AI", profession="student")
```

```python
def student_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

student_details(name="Siddarth", age=25, city="Delhi")
```

```python
def demo(a, *args, **kwargs):
    print(a)
    print(args)
    print(kwargs)

```

```python
demo(10, 20, 30, name="Bipin", role="AI Engineer")
```

### Lambda Functions in Python 
* ðŸ”¹ What is a Lambda Function?

A lambda function is a small, one-line, anonymous function (function without a name).

ðŸ‘‰ Use it when:

Function logic is very small

You need it only once

Clean & short code is preferred

```python
def add(a, b):
    return a + b
y= add(4,10)
y
```

```python
z= lambda a,b:a+b
y = z(4,10)
y
```

```python
def square_add(x):
    y= x*x +100
    return y

sq_result= square_add(10)
sq_result
```

```python
m= lambda x:x*x+100
sq_result= m(10)
sq_result
```

lambda arguments: expression


```python
add = lambda a, b: a + b

add(4,5)
```
