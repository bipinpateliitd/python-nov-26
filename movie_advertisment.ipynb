```python
from dotenv import load_dotenv
load_dotenv(dotenv_path="/home/bipin/Documents/genai/g25-nov-hindi/python-nov/.env")

from langchain_ollama import ChatOllama
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_openai import ChatOpenAI

llm_gpt = ChatOpenAI(model="gpt-5-nano") #openai
llm_google = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite") # google
llm_judge_google = ChatGoogleGenerativeAI(model="gemini-2.5-flash") #google
llm_judge_openai=ChatOpenAI(model="gpt-5-mini") #openai
```

```python
movie_description="""Dhurandhar (2025) is a Hindi-language Indian spy action thriller film directed and co-produced by Aditya Dhar, starring Ranveer Singh, Akshaye Khanna, R. Madhavan, Sanjay Dutt, Arjun Rampal and others. The story follows an Indian intelligence agent embedded deep within Karachi’s dangerous underworld as part of a long-term covert mission to dismantle terror networks and criminal syndicates, blending high-stakes espionage with gritty action. Released on 5 December 2025, the film has become a major box-office success and is noted for its ambitious scale, ensemble cast performances and tense narrative, launching a two-part saga"""
```

```python
input_prompt =f"""Create a catchy promotional advertisement text in Hindi for the movie *Dhurandhar (2025) {movie_description}*.
The ad should be high-energy, engaging, and appealing to Indian audiences.
Include:
- The spy-action thriller theme
- Star cast like Ranveer Singh and Akshaye Khanna
- Elements of patriotism, action, drama
- Important release info (now in theatres / release date)
Make the tone dramatic and cinematic, like a movie promo tagline.
Use short punchy lines and emotional hooks."""
```

```python
print(input_prompt)
```

```python
response_gpt = llm_gpt.invoke(input_prompt)
response_google = llm_google.invoke(input_prompt)
```

```python
print(response_gpt.content)
```

```python
print(response_google.content)
```

```python
judge_prompt = """
You are an expert evaluator judging Hindi movie promotional content
for Indian audiences.

Evaluate each response based on:
1. Catchiness
2. Indian cultural appeal
3. Language quality (Hindi / Hinglish)
4. Creativity
5. Promotional impact

User Prompt:
{question}

Responses:
{responses}

Rules:
- Be unbiased
- Judge ONLY based on quality
- Pick ONE winner

Return output in STRICT JSON only:
{{
  "scores": {{
    "catchiness": number,
    "culture": number,
    "language": number,
    "creativity": number,
    "impact": number,
    "total": number
  }},
  "reason": "short explanation"
}}
"""
```

```python
final_prompt = judge_prompt.format(
    question=input_prompt,
    responses=response_google
)

google_judge_google_lite_evaluation = llm_judge_google.invoke(final_prompt)

```

```python
print(google_judge_google_lite_evaluation.content)
```

```python
final_prompt = judge_prompt.format(
    question=input_prompt,
    responses=response_gpt
)

google_judge_google_gpt_evaluation = llm_judge_google.invoke(final_prompt)

print(google_judge_google_gpt_evaluation.content)
```
