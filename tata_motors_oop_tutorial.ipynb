## 3. Car Features (Instance Methods)

Every Tata car can perform actions like starting, driving, and displaying information.

## 5. Vehicle Hierarchy (Inheritance)

Tata Motors has different categories of vehicles. Let's create a hierarchy:
- Base: TataVehicle
- SUVs: Harrier, Safari
- Compact SUVs: Nexon, Punch
- Electric Vehicles: Nexon EV, Tiago EV

```python
# Parent class - Base vehicle
class TataVehicle:
    """Base class for all Tata Motors vehicles"""
    
    manufacturer = "Tata Motors Limited"
    
    def __init__(self, model, price, engine_type):
        self.model = model
        self.price = price
        self.engine_type = engine_type
        self.mileage = 0
    
    def start_engine(self):
        return f"ğŸš— {self.model} engine started!"
    
    def get_info(self):
        return f"{self.model} - â‚¹{self.price} Lakhs ({self.engine_type})"

# Child class - SUV
class TataSUV(TataVehicle):
    """Tata Motors SUV - Inherits from TataVehicle"""
    
    def __init__(self, model, price, engine_type, seating_capacity, ground_clearance):
        super().__init__(model, price, engine_type)  # Call parent constructor
        self.seating_capacity = seating_capacity
        self.ground_clearance = ground_clearance  # in mm
        self.four_wheel_drive = False
    
    def enable_4wd(self):
        """Enable 4-wheel drive mode"""
        self.four_wheel_drive = True
        return f"ğŸ”ï¸ {self.model} 4WD mode activated! Ready for adventure!"
    
    def get_info(self):
        """Override parent method with SUV-specific info"""
        base_info = super().get_info()
        return f"{base_info} | {self.seating_capacity}-Seater | {self.ground_clearance}mm clearance"

# Child class - Compact SUV
class TataCompactSUV(TataVehicle):
    """Tata Motors Compact SUV"""
    
    def __init__(self, model, price, engine_type, boot_space):
        super().__init__(model, price, engine_type)
        self.boot_space = boot_space  # in liters
    
    def get_info(self):
        base_info = super().get_info()
        return f"{base_info} | Boot Space: {self.boot_space}L"

# Creating different types of vehicles
harrier = TataSUV("Harrier", 26.44, "Diesel", 7, 205)
safari = TataSUV("Safari", 28.06, "Diesel", 7, 205)
nexon = TataCompactSUV("Nexon", 14.99, "Petrol", 350)
punch = TataCompactSUV("Punch", 9.99, "Petrol", 366)

print("ğŸš™ SUVs:")
print(harrier.get_info())
print(safari.get_info())
print(harrier.enable_4wd())
print()
print("ğŸš— Compact SUVs:")
print(nexon.get_info())
print(punch.get_info())
```

## 6. Electric Vehicles (Method Overriding)

Tata's electric vehicles work differently from traditional cars. Let's override methods for EVs.

```python
class TataVehicle:
    """Base Tata Vehicle"""
    
    def __init__(self, model, price):
        self.model = model
        self.price = price
        self.mileage = 0
    
    def start_engine(self):
        return f"ğŸ”¥ {self.model} engine started with ignition!"
    
    def refuel(self, amount):
        return f"â›½ Added {amount} liters of fuel to {self.model}"

class TataElectricVehicle(TataVehicle):
    """Tata Electric Vehicle - Overrides traditional vehicle methods"""
    
    def __init__(self, model, price, battery_capacity, range_km):
        super().__init__(model, price)
        self.battery_capacity = battery_capacity  # in kWh
        self.range_km = range_km
        self.battery_level = 100  # percentage
    
    # Override start_engine for silent electric start
    def start_engine(self):
        return f"âš¡ {self.model} silently powered on! Battery: {self.battery_level}%"
    
    # Override refuel with charging
    def refuel(self, amount):
        return f"âš ï¸ {self.model} is electric! Use charge() method instead."
    
    # New method specific to EVs
    def charge(self, duration_hours, charging_type="Normal"):
        """Charge the electric vehicle"""
        if charging_type == "Fast":
            charge_rate = 30  # 30% per hour
            emoji = "âš¡"
        else:
            charge_rate = 15  # 15% per hour
            emoji = "ğŸ”Œ"
        
        charge_gained = min(duration_hours * charge_rate, 100 - self.battery_level)
        self.battery_level += charge_gained
        
        return f"{emoji} {self.model} charged for {duration_hours}h ({charging_type}). Battery: {self.battery_level}%"
    
    def get_range_info(self):
        """Get current range based on battery level"""
        current_range = (self.battery_level / 100) * self.range_km
        return f"ğŸ”‹ {self.model}: {self.battery_level}% battery | Range: {current_range:.0f} km"

# Traditional vehicle
harrier = TataVehicle("Harrier", 26.44)

# Electric vehicles
nexon_ev = TataElectricVehicle("Nexon EV Max", 18.49, 40.5, 453)
tiago_ev = TataElectricVehicle("Tiago EV", 11.99, 26, 315)

print("ğŸš™ Traditional Vehicle:")
print(harrier.start_engine())
print(harrier.refuel(50))
print()
print("âš¡ Electric Vehicles:")
print(nexon_ev.start_engine())
print(nexon_ev.refuel(50))  # This will show a warning
print(nexon_ev.get_range_info())
print()
print("Simulating battery drain...")
nexon_ev.battery_level = 30
print(nexon_ev.get_range_info())
print(nexon_ev.charge(2, "Fast"))
print(nexon_ev.get_range_info())
```

## 7. Special Features (Magic Methods)

Let's add special methods to compare cars, display them nicely, and perform operations.

```python
class TataMotorsCar:
    """Tata Motors Car with Magic Methods"""
    
    def __init__(self, model, price, engine_cc, mileage_kmpl):
        self.model = model
        self.price = price  # in lakhs
        self.engine_cc = engine_cc
        self.mileage_kmpl = mileage_kmpl
    
    # String representation for print()
    def __str__(self):
        return f"Tata {self.model} (â‚¹{self.price}L)"
    
    # Developer-friendly representation
    def __repr__(self):
        return f"TataMotorsCar('{self.model}', {self.price}, {self.engine_cc}, {self.mileage_kmpl})"
    
    # Comparison: Which car is more expensive?
    def __lt__(self, other):
        return self.price < other.price
    
    def __gt__(self, other):
        return self.price > other.price
    
    def __eq__(self, other):
        return self.price == other.price
    
    # Addition: Calculate total price of multiple cars
    def __add__(self, other):
        return self.price + other.price
    
    # Length: Engine capacity
    def __len__(self):
        return self.engine_cc

# Creating Tata cars
punch = TataMotorsCar("Punch", 9.99, 1199, 18.8)
nexon = TataMotorsCar("Nexon", 14.99, 1497, 17.4)
harrier = TataMotorsCar("Harrier", 26.44, 1956, 16.8)
safari = TataMotorsCar("Safari", 28.06, 1956, 16.1)

print("ğŸš— Using __str__:")
print(punch)
print(nexon)
print()

print("ğŸ”§ Using __repr__:")
print(repr(harrier))
print()

print("ğŸ’° Price Comparisons:")
print(f"Is Punch cheaper than Nexon? {punch < nexon}")
print(f"Is Safari more expensive than Harrier? {safari > harrier}")
print()

print("ğŸª Buying Multiple Cars:")
total_price = punch + nexon + harrier
print(f"Total price for Punch + Nexon + Harrier: â‚¹{total_price} Lakhs")
print()

print("ğŸ”§ Engine Capacity (using __len__):")
print(f"{nexon.model} engine: {len(nexon)} cc")
print(f"{harrier.model} engine: {len(harrier)} cc")
```

## 8. Factory Methods (Class Methods)

Tata Motors has pre-configured popular variants. Let's use class methods as factory methods.

```python
class TataMotorsCar:
    """Tata Motors Car with Factory Methods"""
    
    total_sales = 0
    
    def __init__(self, model, variant, price, color, features):
        self.model = model
        self.variant = variant
        self.price = price
        self.color = color
        self.features = features
        TataMotorsCar.total_sales += 1
    
    # Factory method for popular Nexon variant
    @classmethod
    def nexon_xzplus_petrol(cls):
        """Pre-configured Nexon XZ+ Petrol"""
        features = ["Sunroof", "Wireless Charger", "JBL Sound", "iRA Connected"]
        return cls("Nexon", "XZ+ Petrol", 14.99, "Flame Red", features)
    
    @classmethod
    def nexon_ev_max(cls):
        """Pre-configured Nexon EV Max"""
        features = ["453km Range", "Fast Charging", "Sunroof", "Connected Tech"]
        return cls("Nexon EV", "Max Long Range", 18.49, "Intensi-Teal", features)
    
    @classmethod
    def harrier_dark_edition(cls):
        """Pre-configured Harrier Dark Edition"""
        features = ["Dark Chrome", "Panoramic Sunroof", "JBL", "ADAS Level 2"]
        return cls("Harrier", "XZ+ Dark", 26.44, "Oberon Black", features)
    
    @classmethod
    def punch_adventure(cls):
        """Pre-configured Punch Adventure"""
        features = ["Dual Tone", "Projector Lamps", "iRA", "Cruise Control"]
        return cls("Punch", "Adventure Rhythm", 9.99, "Calypso Red", features)
    
    # Static method for EMI calculation
    @staticmethod
    def calculate_emi(price_lakhs, down_payment_percent, tenure_months, interest_rate):
        """Calculate monthly EMI for car loan"""
        price = price_lakhs * 100000
        down_payment = price * (down_payment_percent / 100)
        loan_amount = price - down_payment
        
        monthly_rate = interest_rate / (12 * 100)
        emi = loan_amount * monthly_rate * ((1 + monthly_rate) ** tenure_months) / \
              (((1 + monthly_rate) ** tenure_months) - 1)
        
        return round(emi, 2)
    
    def display_info(self):
        features_str = ", ".join(self.features)
        return f"""
ğŸš— {self.model} - {self.variant}
   ğŸ’° Price: â‚¹{self.price} Lakhs
   ğŸ¨ Color: {self.color}
   âœ¨ Features: {features_str}
        """

# Using factory methods to create pre-configured cars
car1 = TataMotorsCar.nexon_xzplus_petrol()
car2 = TataMotorsCar.nexon_ev_max()
car3 = TataMotorsCar.harrier_dark_edition()
car4 = TataMotorsCar.punch_adventure()

print("ğŸ­ Pre-configured Tata Motors Cars:")
print(car1.display_info())
print(car2.display_info())
print(car3.display_info())

print(f"\nğŸ“Š Total cars sold: {TataMotorsCar.total_sales}")

# Using static method for EMI calculation
print("\nğŸ’³ EMI Calculator:")
emi = TataMotorsCar.calculate_emi(
    price_lakhs=14.99,
    down_payment_percent=20,
    tenure_months=60,
    interest_rate=8.5
)
print(f"Nexon XZ+ EMI (60 months, 8.5% interest, 20% down): â‚¹{emi:,.2f}/month")
```

## 9. Advanced Features (Properties)

Let's use properties to manage car attributes with validation and computed values.

```python
class TataMotorsCar:
    """Tata Motors Car with Properties"""
    
    def __init__(self, model, price, fuel_capacity):
        self.model = model
        self._price = price  # Private attribute
        self._fuel_capacity = fuel_capacity
        self._fuel_level = fuel_capacity  # Start with full tank
        self._odometer = 0
    
    # Property for price with validation
    @property
    def price(self):
        """Get the car price"""
        return self._price
    
    @price.setter
    def price(self, value):
        """Set price with validation"""
        if value < 0:
            raise ValueError("Price cannot be negative!")
        if value > 100:
            raise ValueError("Price seems too high! Check the value.")
        self._price = value
    
    # Property for fuel level
    @property
    def fuel_level(self):
        """Get current fuel level"""
        return self._fuel_level
    
    @fuel_level.setter
    def fuel_level(self, value):
        """Set fuel level with validation"""
        if value < 0:
            raise ValueError("Fuel level cannot be negative!")
        if value > self._fuel_capacity:
            print(f"âš ï¸ Tank overflow! Setting to max capacity: {self._fuel_capacity}L")
            self._fuel_level = self._fuel_capacity
        else:
            self._fuel_level = value
    
    # Computed property - fuel percentage
    @property
    def fuel_percentage(self):
        """Calculate fuel level as percentage"""
        return (self._fuel_level / self._fuel_capacity) * 100
    
    # Computed property - price in USD
    @property
    def price_usd(self):
        """Convert price to USD (assuming 1 Lakh = 1200 USD)"""
        return self._price * 1200
    
    # Odometer property
    @property
    def odometer(self):
        """Get odometer reading"""
        return self._odometer
    
    def drive(self, distance, fuel_efficiency=15):
        """Drive the car"""
        fuel_needed = distance / fuel_efficiency
        
        if fuel_needed > self._fuel_level:
            return f"âš ï¸ Not enough fuel! Need {fuel_needed:.1f}L, have {self._fuel_level:.1f}L"
        
        self._fuel_level -= fuel_needed
        self._odometer += distance
        return f"ğŸš— Drove {distance}km. Fuel: {self.fuel_percentage:.1f}%, Odometer: {self._odometer}km"
    
    def refuel(self, liters):
        """Refuel the car"""
        self.fuel_level += liters  # Uses the setter with validation
        return f"â›½ Refueled {liters}L. Current fuel: {self.fuel_percentage:.1f}%"

# Using the car with properties
nexon = TataMotorsCar("Nexon", 14.99, 44)

print(f"ğŸš— {nexon.model}")
print(f"ğŸ’° Price: â‚¹{nexon.price} Lakhs (${nexon.price_usd:,.0f})")
print(f"â›½ Fuel: {nexon.fuel_level}L ({nexon.fuel_percentage:.0f}%)")
print()

print(nexon.drive(100))
print(nexon.drive(200))
print()

print(nexon.refuel(20))
print()

# Try to set invalid price
try:
    nexon.price = -5
except ValueError as e:
    print(f"âŒ Error: {e}")

# Try to overfill tank
print(nexon.refuel(50))
```

## 10. Complete Real-World Example: Tata Showroom System

```python
class TataShowroom:
    """Tata Motors Showroom Management System"""
    
    def __init__(self, location):
        self.location = location
        self.inventory = []
        self.sold_cars = []
    
    def add_car(self, car):
        """Add a car to inventory"""
        self.inventory.append(car)
        return f"âœ… {car.model} added to {self.location} showroom"
    
    def sell_car(self, model):
        """Sell a car from inventory"""
        for car in self.inventory:
            if car.model == model:
                self.inventory.remove(car)
                self.sold_cars.append(car)
                return f"ğŸ‰ Sold {car.model} for â‚¹{car.price} Lakhs!"
        return f"âŒ {model} not available in inventory"
    
    def show_inventory(self):
        """Display all available cars"""
        if not self.inventory:
            return "ğŸ“¦ No cars in inventory"
        
        result = f"\nğŸª {self.location} Showroom Inventory:\n"
        result += "=" * 50 + "\n"
        for i, car in enumerate(self.inventory, 1):
            result += f"{i}. {car}\n"
        return result
    
    def get_sales_report(self):
        """Generate sales report"""
        total_revenue = sum(car.price for car in self.sold_cars)
        return f"""
ğŸ“Š Sales Report - {self.location}
{'=' * 40}
Cars Sold: {len(self.sold_cars)}
Total Revenue: â‚¹{total_revenue} Lakhs
Cars in Stock: {len(self.inventory)}
        """

# Create showroom
showroom = TataShowroom("Mumbai")

# Add cars using factory methods
print(showroom.add_car(TataMotorsCar.nexon_xzplus_petrol()))
print(showroom.add_car(TataMotorsCar.nexon_ev_max()))
print(showroom.add_car(TataMotorsCar.harrier_dark_edition()))
print(showroom.add_car(TataMotorsCar.punch_adventure()))

# Show inventory
print(showroom.show_inventory())

# Sell some cars
print("\nğŸ’¼ Sales Transactions:")
print(showroom.sell_car("Nexon"))
print(showroom.sell_car("Harrier"))

# Updated inventory
print(showroom.show_inventory())

# Sales report
print(showroom.get_sales_report())
```

```python
# Solution 1: Tata Truck Class
class TataVehicle:
    def __init__(self, model, price):
        self.model = model
        self.price = price

class TataTruck(TataVehicle):
    def __init__(self, model, price, load_capacity, num_wheels):
        super().__init__(model, price)
        self.load_capacity = load_capacity  # in tons
        self.num_wheels = num_wheels
        self.current_load = 0
    
    def load_cargo(self, weight):
        if self.current_load + weight > self.load_capacity:
            return f"âš ï¸ Overload! Capacity: {self.load_capacity}T, Current: {self.current_load}T, Trying to add: {weight}T"
        self.current_load += weight
        return f"âœ… Loaded {weight}T. Current load: {self.current_load}T / {self.load_capacity}T"
    
    def unload_cargo(self):
        unloaded = self.current_load
        self.current_load = 0
        return f"ğŸ“¦ Unloaded {unloaded}T. Truck is now empty."
    
    def get_load_status(self):
        percentage = (self.current_load / self.load_capacity) * 100
        return f"ğŸš› {self.model}: {self.current_load}T / {self.load_capacity}T ({percentage:.1f}%)"

# Test
truck = TataTruck("Tata Prima", 35, 25, 10)
print(truck.load_cargo(15))
print(truck.load_cargo(8))
print(truck.get_load_status())
print(truck.load_cargo(5))  # Overload
print(truck.unload_cargo())
```

```python
# Solution 2: Service Center Class
class TataServiceCenter:
    def __init__(self, location):
        self.location = location
        self.cars_in_service = []
        self.service_costs = {
            "Basic": 3000,
            "Standard": 6000,
            "Premium": 10000
        }
    
    def accept_car(self, car):
        self.cars_in_service.append(car)
        return f"ğŸ”§ {car.model} accepted for service at {self.location}"
    
    def complete_service(self, car_model):
        for car in self.cars_in_service:
            if car.model == car_model:
                self.cars_in_service.remove(car)
                return f"âœ… {car_model} service completed!"
        return f"âŒ {car_model} not found in service"
    
    def calculate_service_cost(self, car, service_type):
        if service_type not in self.service_costs:
            return "âŒ Invalid service type"
        
        cost = self.service_costs[service_type]
        return f"ğŸ’° {service_type} service for {car.model}: â‚¹{cost}"

# Test
service_center = TataServiceCenter("Delhi")
nexon = TataMotorsCar.nexon_xzplus_petrol()
print(service_center.accept_car(nexon))
print(service_center.calculate_service_cost(nexon, "Standard"))
print(service_center.complete_service("Nexon"))
```
