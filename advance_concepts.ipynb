*  Encapsulation means binding data (variables) and methods (functions) together inside a class
* controlling direct access to that data.
- Protects data from accidental modification

- Python does encapsulation using:

- Public members

- Protected members (_)

- Private members (__)

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance  # public

```

```python
#create object

acc = BankAccount(1000)
```

```python
class BankAccount:
    def __init__(self, balance, secret_name, password):
        self.balance = balance  # public
        self._secret_name= secret_name# protected
        self.__password= password # private
    def password_name(self):
        return f" the password is {self.__password}"
```

```python
#Protected Members (_variable)
class BankAccount:
    def __init__(self, balance):
        self._balance = balance  # protected

acc = BankAccount(1000)
print(acc._balance)  # ‚ö†Ô∏è possible, but not recommended
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

acc = BankAccount(1000)

acc.get_balance()
```

```python
acc.deposit(2000)
```

```python
acc.get_balance()
```

```python
class User:
    def __init__(self, password):
        self.__password = password  # üîê private

    def check_password(self, input_password):
        return self.__password == input_password

    def change_password(self, old_password, new_password):
        if self.__password == old_password:
            self.__password = new_password
            print("Password changed successfully")
        else:
            print("Wrong old password")
```

```python
class LoginService:
    def login(self, username, password):
        if self.__validate(username, password):
            print("Login successful")
        else:
            print("Invalid credentials")

    def __validate(self, username, password):  # encapsulated method
        return username == "admin" and password == "1234"
```

```python
l = LoginService()
l.login("admin", "1234")   # ‚úÖ
# l.__validate(...)        # ‚ùå not allowed
```

# Inheritance
- Inheritance means one class (child) acquires properties and methods of another class (parent)

```python
class Vehicle: # parent class
    def start(self):
        print("Vehicle started")
```

```python
a1= Vehicle()
```

```python
a1.start()
```

```python

class Car(Vehicle):   # Car inherits Vehicle
    def drive(self):
        print("Car is driving")
```

```python
a2= Car()
```

```python
a2.drive()
```

```python
a2.start()
```

```python

class Car(Vehicle):   # Car inherits Vehicle
    def drive(self):
        print("Car is driving")

c = Car()
c.start()   # inherited method
c.drive()
```

```python
class Vehicle: # parent class
    def __init__(self, brand):
        self.brand = brand

```

```python
a1= Vehicle("Hyundai")
a1.brand
```

```python
class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)   # call parent constructor
        self.model = model

    def info(self):
        print(self.brand, self.model)
```

```python
a2.brand
```

```python
a2.model
```

```python
class Vehicle:
    def __init__(self, brand, fuel_type, wheels):
        self.brand = brand
        self.fuel_type = fuel_type
        self.wheels = wheels

    def vehicle_info(self):
        print(f"Brand: {self.brand}")
        print(f"Fuel: {self.fuel_type}")
        print(f"Wheels: {self.wheels}")
```

```python
class Car(Vehicle):
    def __init__(self, brand, fuel_type, wheels, model, price):
        super().__init__(brand, fuel_type, wheels)  # üëà pass parent args
        self.model = model
        self.price = price

    def car_info(self):
        self.vehicle_info()
        print(f"Model: {self.model}")
        print(f"Price: {self.price}")
```

```python
c = Car("Tata","Petrol","4", "Nexon","11.6L")
```

```python
c.vehicle_info()
```

```python
c.car_info()
```

```python
class Vehicle:
    def __init__(self, brand, fuel, wheels, max_speed):
        self.brand = brand
        self.fuel = fuel
        self.wheels = wheels
        self.max_speed = max_speed
```

```python
class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand, "Petrol", 4, 180)  # fixed values
        self.model = model
```

- All required parent constructor arguments must be passed to super().__init__(), either explicitly, via defaults, or by being set by the child  class.

1Ô∏è‚É£ Single Inheritance

One parent ‚Üí one child

```python
# single inheritance
class A:
    def show(self):
        print("Class A")

class B(A):
    pass
```

```python
class User:
    def login(self):
        print("User logged in")

class Admin(User):
    def delete_user(self):
        print("User deleted")

a = Admin()
a.login()
a.delete_user()
```

2Ô∏è‚É£ Multilevel Inheritance

Grandparent ‚Üí parent ‚Üí child

```python
class Vehicle:
    def start(self):
        print("Vehicle started")

class Car(Vehicle):
    pass

class ElectricCar(Car):
    def charge(self):
        print("Charging")

e = ElectricCar()
e.start()
e.charge()
```

```python
class Device: # grandparent
    def power_on(self): #property-power_on
        print("Device ON")

class Phone(Device):#parent
    def call(self): #property-power_on, call
        print("Calling")

class SmartPhone(Phone):#child
    def browse(self):#property-power_on, call, browse
        print("Browsing internet")

s = SmartPhone()
s.power_on()
s.call()
s.browse()
```

3Ô∏è‚É£ Multiple Inheritance

One child, multiple parents

```python
class Engine:#parent1
    def engine_info(self):
        print("Engine")

class MusicSystem:#parent2
    def music(self):
        print("Music system")

class Car(Engine, MusicSystem):#child
    pass
```

```python
class Camera:
    def capture(self):
        print("Capturing image")

class GPS:
    def navigate(self):
        print("Navigating")

class SmartCar(Camera, GPS):
    pass

c = SmartCar()
c.capture()
c.navigate()
```

4Ô∏è‚É£ Hierarchical Inheritance

One parent, many children

```python
class Vehicle:
    pass

class Car(Vehicle):
    pass

class Bike(Vehicle):
    pass
```

```python
class Employee:
    def work(self):
        print("Working")

class Developer(Employee):
    def code(self):
        print("Writing code")

class Manager(Employee):
    def manage(self):
        print("Managing team")
```

5Ô∏è‚É£ Hybrid Inheritance

Combination of multiple types

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass
```

```python
class Account:
    def open(self):
        print("Account opened")

class Savings(Account):
    def save(self):
        print("Saving money")

class Current(Account):
    def transact(self):
        print("Business transaction")

class PremiumAccount(Savings, Current):
    pass

p = PremiumAccount()
p.open()
p.save()
p.transact()
```

- Method overriding happens when a child class provides its own implementation of a method that already exists in the parent class, with the same method name.

```python
class Vehicle:
    def start(self):
        print("Vehicle started")

class Car(Vehicle):
    def start(self):        # overridden method
        print("Car started")

```

```python

v = Vehicle()
c = Car()

```

```python
v.start()
```

```python
c.start()
```

```python
class User:
    def access(self):
        print("Basic user access")

class Admin(User):
    def access(self):
        print("Admin full access")

u = User()
a = Admin()

u.access()
a.access()
```
