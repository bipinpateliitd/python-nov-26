## 1. Introduction to Classes and Objects 
**Class**: A blueprint for creating objects. It defines attributes and methods. 

**Object**: An instance of a class. It's a concrete entity based on the class blueprint.

ðŸ”° LEVEL 1: SIMPLE CLASS (No attributes, no methods)

```python
class TataCar:
    pass
```

```python
car1 = TataCar()
car2 = TataCar()

print(car1)
print(car2)
```

ðŸŸ¡ LEVEL 2: CLASS WITH ATTRIBUTES (Data add )

```python
class TataCar:
    def __init__(self, model, color):
        self.model = model
        self.color = color
```

```python
car1 = TataCar("Nexon", "Red")
car2 = TataCar("Punch", "Blue")

```

ðŸŸ¢ LEVEL 3: ATTRIBUTES + METHODS (Car ka kaam)

```python
class TataCar:
    def __init__(self, model, color):
        self.model = model
        self.color = color

    def start(self):
        print(self.model, "is starting")

    def stop(self):
        print(self.model, "has stopped")
```

```python
car1 = TataCar("Nexon", "Red")

car1.start()
car1.stop()
```

ðŸ”µ LEVEL 4: MORE ATTRIBUTES + DECISION LOGIC

```python
class TataCar:
    def __init__(self, model, price):
        self.model = model
        self.price = price

    def category(self):
        if self.price > 12_00_000:
            print(self.model, "is a premium car")
        else:
            print(self.model, "is a budget car")
```

```python
car1 = TataCar("Nexon", 9_00_000)
car2 = TataCar("Harrier", 15_00_000)

car1.category()
car2.category()
```

ðŸŸ£ LEVEL 5: ADVANCED (Class variable + Instance variable)

- Company-level + car-level data

```python
class TataCar:
    company = "Tata Motors"   # class variable

    def __init__(self, model, price):
        self.model = model    # instance variable
        self.price = price

    def show_details(self):
        print(self.company, "-", self.model, "-", self.price)
```

```python
car1 = TataCar("Punch", 6_00_000)
car2 = TataCar("Safari", 18_00_000)

car1.show_details()
car2.show_details()
```

ðŸ”´ LEVEL 6: REAL-WORLD STYLE (Multiple methods)

```python
class TataCar:
    company = "Tata Motors"

    def __init__(self, model, fuel):
        self.model = model
        self.fuel = fuel

    def start(self):
        print(self.model, "started using", self.fuel)

    def mileage(self):
        if self.fuel == "Diesel":
            print("High mileage")
        else:
            print("Average mileage")
```

```python
car1 = TataCar("Altroz", "Petrol")
car2 = TataCar("Safari", "Diesel")

car1.start()
car1.mileage()

car2.start()
car2.mileage()
```
