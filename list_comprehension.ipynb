ðŸ¤” What is List Comprehension?

List comprehension is a short and clean way to create a new list from:

another list

a range

or any iterable

ðŸ‘‰ It replaces loops + append() in one line.

```python
[new_value for item in iterable]
```

```python
numbers = [1, 5, 9, 4] #iterable
squares = []
for x in numbers:
    sq= x*x
    squares.append(sq)
```

```python
squares
```

```python

#result= [new_value for item in iterable]


sqr = (x*x for x in numbers)
sqr
```

```python
#result= [new_value for item in iterable]


nums = [ x**3 for x in range(1, 9)]
nums
```

```python
num = [1, 2, 3, 4, 5, 6, 34, 26, 11, 99] #findout even number  using list comprehension

#result= [new_value for item in iterable]

even_num =[y for y in num if y%2==0]


```

```python
even_num
```

```python
marks = [60, 85, 90, 70, 95]

top_marks = [m for m in marks if m > 80]
print(top_marks)
```

```python
numbers = [1, 2, 3, 4]

result = ["Even" if x % 2 == 0 else "Odd" for x in numbers]
print(result)
```
